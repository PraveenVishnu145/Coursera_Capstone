```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',skiprows=0)[0]
indexNames = data[ (data['Borough'] == 'Not assigned') & (data['Neighbourhood'] == 'Not assigned') ].index
# Delete these row indexes from dataFrame
data.drop(indexNames , inplace=True)
for index, row in data.iterrows():
        if row['Neighbourhood'] == 'Not assigned':
            data.set_value(index, 'Neighbourhood', row['Borough'])
data2 = data.groupby("Postcode")['Neighbourhood'].apply(lambda tags: ','.join(tags))
for index, row in data.iterrows():
    data.set_value(index, 'Neighbourhood', data2[row['Postcode']])
df = data.drop_duplicates(keep='last')
df.head()
```

```python
df.shape
```

```python
coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
coordinates.head()
```

```python
merged_inner = pd.merge(left=coordinates, right=df, left_on='Postal Code', right_on='Postcode')
final_df = merged_inner[['Postal Code','Borough','Neighbourhood','Latitude','Longitude']]
final_df.head()
```

```python
final_df_Toronto =  final_df[final_df['Borough'].str.contains("Toronto")]
final_df_Toronto_index = final_df_Toronto.set_index('Postal Code')
final_df_Toronto_index.head(38)
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(final_df_Toronto_index['Latitude'], final_df_Toronto_index['Longitude'], final_df_Toronto_index['Borough'], final_df_Toronto_index['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
